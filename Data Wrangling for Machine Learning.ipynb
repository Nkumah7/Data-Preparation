```python
import sklearn
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
# To load data set
automobile = pd.read_csv('Automobile Price Data.csv')

# To see what the data set looks like
automobile.head(5)
```

```python
# To check the shape of the data set

automobile.shape
```

```python
# To replace '?' with NaN in data set
automobile = automobile.replace('?', np.nan)

# To check updated dataset
automobile
```

```python
# Evaluate the number of missing data in each column

automobile.isna().sum()
```

```python
# Check data type

automobile.dtypes
```

```python
# Calculate mean of "normalized-losses"

mean_of_loss = automobile["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", mean_of_loss)
```

```python
# Replace the missing data in "normalized-losses" column with mean_of_loss

automobile['normalized-losses'].replace(np.nan, mean_of_loss, inplace = True)
```

```python
automobile.isna().sum()
```

```python
# Calculate mean of "bore"

mean_of_bore = automobile["bore"].astype("float").mean(axis=0)
print("Average of normalized-losses:", mean_of_bore)
```

```python
# Replace the missing data in "bore" column with mean_of_bore

automobile["bore"].replace(np.nan, mean_of_bore, inplace = True)
automobile.isna().sum()
```

```python
# Calculate mean of "stroke"

mean_of_stroke = automobile['stroke'].astype("float").mean(axis=0)
print("Mean of Stroke:", mean_of_stroke)
```

```python
# Replace the missing data in "stroke" column with mean_of_stroke

automobile["stroke"].replace(np.nan, mean_of_stroke, inplace = True)
automobile.isna().sum()
```

```python
# Calculate mean of "horsepower"

mean_of_horsepower = automobile["horsepower"].astype("float").mean(axis=0)
print("Mean of horsepower:", mean_of_horsepower)
```

```python
# Replace the missing data in "horsepower" column with mean_of_horsepower

automobile["horsepower"].replace(np.nan, mean_of_horsepower, inplace = True)
automobile.isna().sum()
```

```python
# Calculate mean of "peak-rpm"

mean_of_peak = automobile["peak-rpm"].astype("float").mean(axis=0)
print("Mean of Peak-rpm:", mean_of_peak)
```

```python
# Replace the missing data in "peak-rpm" column with mean_of_peak

automobile["peak-rpm"].replace(np.nan, mean_of_peak, inplace = True)
automobile.isna().sum()
```

```python
# Evalaute the value with the most frequency

automobile["num-of-doors"].value_counts().idxmax()
```

```python
# Replace the missing data in "num-of-doors" column with the most frequent value "four"

automobile["num-of-doors"].replace(np.nan, "four", inplace = True)
automobile.isna().sum()
```

```python
# simply drop whole row with NaN in "price" column
automobile.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
automobile.reset_index(drop=True, inplace=True)

automobile.isna().sum()
```

```python
# Drop columns with string not needed for ML

automobile.drop(["make", "fuel-type", "aspiration", "body-style", "drive-wheels", "engine-location", "engine-type", "fuel-system"], inplace = True, axis = 1)
automobile.head(5)
```

```python
automobile.drop(["num-of-doors", "num-of-cylinders"], inplace = True, axis = 1)
automobile.head(5)
```

```python
automobile.drop(["bore", "stroke", "compression-ratio"], inplace = True, axis = 1)
automobile.head()
```

```python
automobile.dtypes
```

```python
# Convert data types to proper format

automobile["normalized-losses"] = pd.to_numeric(automobile["normalized-losses"], errors = 'coerce')
automobile["horsepower"] = pd.to_numeric(automobile["horsepower"], errors = 'coerce')
automobile["peak-rpm"] = pd.to_numeric(automobile["peak-rpm"], errors = 'coerce')
automobile["price"] = pd.to_numeric(automobile["price"], errors = 'coerce')
automobile.dtypes
```

```python
# Descriptive statistics of the data set

automobile.describe()
```

```python
# Save the cleaned dataset in csv format

automobile.to_csv('new dataset.csv', index = False)
```
