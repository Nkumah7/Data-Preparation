```python
import sklearn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data set
sales = pd.read_csv('advertising.csv')

# View data set
sales.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Set predictor and target to be trained and tested
X = sales[['TV']] #predictor
Y = sales['Sales'] #target

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

# Train Model
linear_model1 = LinearRegression(normalize=True).fit(x_train, y_train)

# Print Training Score
print('Training Score: ', linear_model1.score(x_train, y_train))

# Test Model
y_pred1 = linear_model1.predict(x_test)

# Print Testing Score
print('Testing Score: ', r2_score(y_test, y_pred1))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred1, color='g')

plt.xlabel('TV')
plt.ylabel('Sales')
```

```python
# Set predictor and target data to be trained and tested
X = sales[['Radio']]
Y = sales['Sales']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

# Train Model
linear_model2 = LinearRegression(normalize=True).fit(x_train, y_train)

# Print Training Score
print('Training Score: ', linear_model2.score(x_train, y_train))

# Test Model
y_pred2 = linear_model2.predict(x_test)

# Print Testing Score
print('Testing Score: ', r2_score(y_test, y_pred2))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred2, color='r')

plt.xlabel('Radio')
plt.ylabel('Sales')
```

```python
# Set predictor and target data to be trained and tested
X = sales[['Newspaper']]
Y = sales['Sales']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

# Train the model
linear_model3 = LinearRegression(normalize=True).fit(x_train, y_train)

# Print the Training Score
print('Training Score: ', linear_model3.score(x_train, y_train))

# Test the model
y_pred3 = linear_model3.predict(x_test)

# Print the Testing Score
print('Testing Score: ', r2_score(y_test, y_pred3))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred3)

plt.xlabel('Newspaper')
plt.ylabel('Sales')
```

```python
# Set predictors and target to be trained and tested for linear regression multiple feature
X = sales[['TV', 'Radio', 'Newspaper']]
Y = sales['Sales']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

# Train the model
linear_model4 = LinearRegression(normalize=True).fit(x_train, y_train)

# Print the Training Score
print('Training Score: ', linear_model4.score(x_train, y_train))

#Test the model
y_pred4 = linear_model4.predict(x_test)

# Print the Testing Score

print('Testing Score: ', r2_score(y_test, y_pred4))
```

```python
# Plot line graph
plt.figure(figsize=(20, 10))

plt.plot(y_pred4, label='Predicted')
plt.plot(y_test.values, label='Actual')

plt.ylabel('Sales')

plt.legend()
```
